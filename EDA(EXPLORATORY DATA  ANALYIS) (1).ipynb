# Superstore Dataset: Data Analysis Report 

Introduction:

The Superstore dataset is a comprehensive retail dataset that captures sales transactions, containing information about orders, customers, products, and shipping details across different regions. It provides valuable insights into various business aspects including sales performance, shipping preferences, and customer segmentation, making it an excellent resource for analyzing retail operations and understanding business patterns from 2014 to 2017.

# Data Overview: 

* The dataset includes the following columns:

  
  Order Details: Order_ID,Order_Date,_Ship Date.

  
  Customer Information: Cutomer_ID,Customer_Name.

 
  Product Details: Product_ID,Product_Name

 
  Segmentation: Segment,Category,sub-Category

  
  Geographical Information: Region,Country,City,State

 
  Financial Metrics:Sales,Quantity,Discount,Profit

 

# Data Visualization:

3. Color schemes:
* winter palette for some visualizations
* default color schemes for others
4. proper titles and labels for better understanding
5. Legend usage where multiple categories are shown

These Visualizations help in understanding the distribution and relationships between different variables in the superstore dataset,making it easier to identify patterns and trends in the data.

# Insights:
The Superstore dataset reveals standard class as the most popular shipping mode,indicating customer preference for economical options.Office supplies is the top  selling category,with diverse sub categories across all product lines.There's a clear upward trend in order volumne from 2014-2017,suggesting consistent business growth and expanding customer base across different segments and regions.

# SHIP MODE

```python
dataset['Ship Mode'].value_counts()
```

```python
x=dataset['Ship Mode'].value_counts().index
y=dataset['Ship Mode'].value_counts().values
```

```python
plt.pie(y,labels=x,startangle=60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
sns.countplot(x="Ship Mode",
              data=dataset,
              palette="winter")
plt.title("Count Of Ship Mode")
plt.show()
```

#  CATEGORY

```python
sns.countplot(x="Ship Mode",data=dataset,hue="Category")
plt.title("CATEGORY")
plt.savefig("CATEGORY.jpg")
plt.show()
```

# SEGMENT

```python
sns.countplot(x="Segment",data=dataset)
plt.title("Segment")
plt.show()
```

# SUB CATEGORY

```python
sns.countplot(x="Category",
              data=dataset[dataset["Category"]=="Office Supplies"],
              hue="Sub-Category")
plt.title("SUB CATEGORY")
plt.show()
```

```python
sns.countplot(x="Category",
              data=dataset[dataset["Category"]=="Technology"],
              hue="Sub-Category")
plt.title("SUB CATEGORY")
plt.show()
```

```python
sns.countplot(x="Category",
              data=dataset[dataset["Category"]=="Furniture"],
              hue="Sub-Category")
plt.title("SUB CATEGORY")
plt.show()
```

# ORDER YEAR

```python
 dataset["order year"]= dataset["Order Date"].dt.year
```

```python
dataset["order year"].value_counts()
```

```python
plt.figure(figsize=(10, 6))
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])
dataset.set_index('Order Date', inplace=True)

monthly_sales = dataset['Sales'].resample('M').sum()

plt.figure(figsize=(14, 7))
monthly_sales.plot()
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

# DISCOUNT

```python
plt.figure(figsize=(10, 6))
sns.histplot(dataset['Discount'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Discounts')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```

# PROFIT

```python
sns.barplot(x="Category",y="Profit",data=dataset,estimator="sum")
plt.show()
```

# CITY

```python
dataset['City'].value_counts()[:5]
```

```python
plt.figure(figsize=(10, 6))
numeric_data = dataset.select_dtypes(include=['float64', 'int64'])
corr = numeric_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
